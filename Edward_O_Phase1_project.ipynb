## # Return on investment and rank movies according to return on investment

First create ROI based on total gross

```python
# creates column 'ROI' in the 'movies_budget' data frame which represents the return on investment (ROI) as a percentage of the production budget.
# Used str.replace() method to remove the commas and dollar signs from the 'production_budget' column before converting the resulting string values to float using the astype() method.
movies_budget['ROI'] = ((movies_budget['Total_Gross'] - movies_budget['production_budget'].str.replace(',', '').str.replace('$', '').astype(float)) / movies_budget['production_budget'].str.replace(',', '').str.replace('$', '').astype(float)) * 100
```

```python
movies_budget.info()
```

```python
# convert the columns into floats before attempting to convert
movies_budget['production_budget'] = movies_budget['production_budget'].str.replace(',', '').str.replace('$', '').astype(float)
movies_budget['domestic_gross'] = movies_budget['domestic_gross'].str.replace(',', '').str.replace('$', '').astype(float)
movies_budget['ROI_domestic'] = ((movies_budget['domestic_gross'] - movies_budget['production_budget']) / movies_budget['production_budget']) * 100
```

```python
def convert_rating(rating):
    if isinstance(rating, str) and '/' in rating:
        try:
            return float(rating.split('/')[0]) / float(rating.split('/')[1])
        except:
            return np.nan
    else:
        return np.nan

movies_review['rating'] = movies_review['rating'].apply(convert_rating)
```
